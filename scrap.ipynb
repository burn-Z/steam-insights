```python
import pandas as pd
from os import path, listdir
```

```python
dir= 'data/'
file_names = ['categories.zip', 'descriptions.zip','games.zip','genres.zip','promotional.zip','reviews.zip','steamspy_insights.zip','tags.zip']

for file in listdir(dir):
    file_path= path.join(dir, file)
    if path.isfile(file_path):
        try:
            temp_df = pd.read_csv(dir + file, compression= 'zip', engine= 'pyarrow')
            print(f"Successfully read {file} with shape {temp_df.shape}")
        except Exception as e:
            print(f"Failed to read {file}: {e}")
        print('*'*50)
```
